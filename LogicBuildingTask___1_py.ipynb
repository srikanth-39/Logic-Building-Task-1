```python
# Given marks in the Problem Statement
marks = [45, 78, 90, 33, 60]

passCount = 0
failCount = 0

# Analyze the results from the given marks
for mark in marks:
    if mark >= 50:
        passCount += 1
    else:
        failCount += 1

# Print the results
print("Total Students:", len(marks))
print("Number of Pass Students:", passCount)
print("Number of Fail Students:", failCount)
```

```python
# Original list
names = [" Alice ", "bob", " CHARLIE "]

# Clean the names
cleaned_names = []
for name in names:
    cleaned_names.append(name.strip().lower())
# Print result
print(cleaned_names)
```

```python
# List of messages
messages = ["Hi", "Welcome to the platform", "OK"]

# Analyze each message from the list
for msg in messages:
    length = len(msg)
    print(f"Message: '{msg}' → Length: {length}")

    if length > 10:
        print("  ⚠️ Message is longer than 10 characters")
```
